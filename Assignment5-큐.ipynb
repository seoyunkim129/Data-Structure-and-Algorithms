<span style="color:red">isQueueFull( ), isQueueEmpty( ), enQueue( ), deQueue( ), peek( ) 함수</span>를 완성시키세요.  
단, 순차 큐(Sequential Queue)를 이용하여 문제를 해결하세요.

```python
def isQueueFull():
    global SIZE, queue, front, rear
    if rear == SIZE - 1:
        return True
    elif front == -1 and rear == SIZE - 1:
        return True
    else:
        return False

def moveQueue():
    global SIZE, queue, front, rear
    for i in range(front + 1, SIZE):
        queue[i - (front + 1)] = queue[i]
        queue[i] = None
    front -= 1
    rear -= 1

def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print("큐가 꽉찼습니다")
        return

    rear += 1
    queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print("큐가 비었습니다")
        return None

    data = queue[front + 1]
    queue[front + 1] = None
    front += 1
    return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print("큐가 비었습니다")
        return None

    return queue[front + 1]

# 전역 변수 선언 부분
SIZE = 5
queue = [None for _ in range(SIZE)]
front = -1
rear = -1

# 메인 코드 부분
if __name__ == "__main__":
    enQueue("코고로")     #  567표
    enQueue("란")         # 1234표
    enQueue("키드")       # 1412표
    enQueue("코난")       # 1994표
    enQueue("하이바라")   # 4869표

    print("2023년 팬들이 뽑은 명탐정 코난 캐릭터 인기순위!")

    ranking = ["5위", "4위", "3위", "2위", "1위"]
    for j in range(rear + 1):
        print(ranking[j], ":", deQueue())
        moveQueue()
        print("      남은 후보:", queue)

    print("발표 종료!")
```

### 문제2)

Assignment5.pdf 파일의 <문제2>의 내용 및 실행결과를 참고하여 아래 함수 선언 부분에서 미완성된  
<span style="color:red">isQueueFull( ), isQueueEmpty( ), enQueue( ), deQueue( ), peek( ) 함수</span>를 완성시키세요.  
단, 원형 큐(Circular Queue)를 이용하여 문제를 해결하세요.

```python
def isQueueFull():
    global SIZE, queue, front, rear
    if (rear+1) % SIZE == front:
        return True
    else:
        return False

def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print("Queue is full!")
        return
    rear = (rear+1) % SIZE
    queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print("Queue is empty!")
        return None
    front = (front+1) % SIZE
    data = queue[front]
    queue[front] = None
    return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print("Queue is empty!")
        return None
    return queue[(front+1) % SIZE]


def calcTime():
    global SIZE, queue, front, rear
    timeSum = 0
    for j in range(front+1, rear+1):
        timeSum = timeSum + queue[j][1]

    return timeSum

# 전역 변수 선언 부분
SIZE = 6
queue = [None for _ in range(SIZE)]
front = 0
rear = 0

# 메인 코드 부분
if __name__ == "__main__":
    waitCall = [("사용", 9), ("고장", 3), ("환불", 4), ("환불", 4), ("고장", 3)]

    for call in waitCall:
        enQueue(call)
        print("귀하의 대기 예상시간은", calcTime(), "분입니다.")
        print("현재 대기 콜 -->", queue)


    print()
    print("최종 대기 콜 -->", queue)
    print("프로그램 종료!")
```
